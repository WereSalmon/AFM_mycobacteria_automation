## Initializing packages, functions

```python
import numpy as np
import os, re, cv2, sknw
from networkx import shortest_path
from networkx import number_connected_components
from scipy.signal import argrelmax
from scipy.interpolate import splprep
from scipy.interpolate import splev
from cellpose import utils
from skimage.morphology import skeletonize
import warnings
import copy
import tools
from create_ridgeline import endpoints
from create_ridgeline import order_by_NN
from create_ridgeline import padskel
from create_ridgeline import longest_path
from create_ridgeline import ridgeline
from create_ridgeline import find_spline
from create_ridgeline import spline_to_img

from tools import masks_list
import matplotlib.pyplot as plt

segments_path = "cellpose_output/"
```

## Initialize data

```python
# Get the filenames for our segmentation datasets. 
segmented = os.listdir(segments_path)
for fichier in segmented[:]: 
    if not(fichier.endswith(".npy")):
        segmented.remove(fichier)

# Sort the filenames by timepoint. 
segmented.sort(key = tools.natural_keys)

# Create a list that records the timepoints for every image. 
time_list = []
for segment in segmented:
    time_list.append(int(re.findall(r'\d+', segment)[0]))

# Initialize lists.
outl_temp = []
maskl_temp = []
iml_temp = []
gradl_temp = []

# Fill lists with img and outline data from each segment
for segment, time in zip(segmented, time_list):
    # Load up the segmentation data. 
    dat = np.load(segments_path + segment, allow_pickle=True).item()
    
    # Remove small masks from the data
    dat["masks"] = utils.fill_holes_and_remove_small_masks(dat["masks"], min_size = 800)
    
    # Populate list with each cell outline.
    outl = utils.outlines_list(dat['masks'])
    outl_temp.append(outl)
    
    # Populate list with each original mask.
    maskl = masks_list(dat['masks'])
    maskl_temp.append(maskl)
    
    # Populate list with flow data
    gradl = dat['flows']
    gradl_temp.append(gradl)
    
    #Populate list with each original image
    iml = dat['img']
    iml_temp.append(iml)
```

## Initialize individual cell

```python
r=22
t=1

# cellpose outputs
mask = (copy.deepcopy(maskl_temp[r][t])>0).astype(np.uint8)
outline = (copy.deepcopy(outl_temp[r][t]))

#find the center of the cell
out = np.transpose(outline)
center = np.array([np.sum(out[0]),np.sum(out[1])])/len(out[0])

# find endpoints
lp,rp = endpoints(outline,center)

# find topological skeleton
skel = padskel(mask)
```

```python
plt.imshow((1-skel)*mask,cmap='gray')
plt.plot(lp[0],lp[1],'r*',markersize=24)
plt.plot(rp[0],rp[1],'r*',markersize = 24)
plt.xlim([35,185])
plt.ylim([145,85])
plt.axis('off')
```

## Divide the skeleton in two

```python
d = np.floor(center[0]).astype(np.uint8)
(m,n)=np.shape(mask)
sides = np.hstack([np.ones((m,d+1))*(-1),np.ones((m,n-d-1))])
plt.imshow(sides*mask*(1-skel), cmap = 'bwr')
plt.plot(lp[0],lp[1],'k*',markersize=24)
plt.plot(rp[0],rp[1],'k*',markersize=24)
plt.plot(center[0],center[1],'k.',markersize=24)
plt.xlim([35,185])
plt.ylim([145,85])
plt.axis('off')
```

## Find the longest path through the skeleton

```python
graph = sknw.build_sknw(skel) #creating the graph from the skeleton
nodes = graph.nodes()
endleft = [i for i in nodes if nodes[i]['o'][1]<=d]
endright = [i for i in nodes if nodes[i]['o'][1]>d]
```

```python
path,start_node,end_node = longest_path(graph,endleft,endright)
```

```python
plt.imshow(sides*mask*(1-skel), cmap = 'bwr')
plt.plot(np.transpose(path)[1],np.transpose(path)[0],'k.')
plt.xlim([35,185])
plt.ylim([145,85])
plt.axis('off')
```

## Ensure the path is properly ordered

```python
path = order_by_NN(path,nodes[start_node]['o'],nodes[end_node]['o'])
```

## Crop 10% off the end of the paths

```python
crop_10 = np.floor(len(path)*0.1).astype(int)
```

```python
path_cropped = copy.deepcopy(path[crop_10:-crop_10])
```

```python
plt.imshow(sides*mask*(1-skel), cmap = 'bwr')
plt.plot(np.transpose(path_cropped)[1],np.transpose(path_cropped)[0],'k.')
plt.xlim([35,185])
plt.ylim([145,85])
plt.axis('off')
```

## Find the spline of from the path 

```python
ridge_path = [np.flip(lp)]+path_cropped+[np.flip(rp)]
[xs,ys],s, max_k = find_spline(ridge_path)
```

```python
plt.imshow(sides*mask, cmap = 'bwr')
plt.plot(xs,ys,'k-',linewidth=2)
plt.plot(lp[0],lp[1],'k*',markersize=24)
plt.plot(rp[0],rp[1],'k*',markersize=24)
plt.xlim([35,185])
plt.ylim([145,85])
plt.axis('off')
```

## Convert the spline to a binary image

```python
ridge = spline_to_img(np.transpose([xs,ys]),mask)
plt.imshow((1-ridge)*mask,cmap='gray')
plt.xlim([35,185])
plt.ylim([145,85])
plt.axis('off')
```
